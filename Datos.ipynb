```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
df = pd.read_excel('registro.xlsx')
```

```python
df.head()
```

```python
telefonos= df['telefono_wa']
```

```python
etiquetas = df["telefono_wa"].unique()

fig, ax= plt.subplots()

plt.pie(telefonos.value_counts(),labels=etiquetas, autopct="%0.2f %%")

ax.set_title("Porcentaje de uso de chatbot")

plt.show()
```
